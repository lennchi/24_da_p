```python
import pandas as pd
path = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/ceska-jmena/jmena.csv"
data = pd.read_csv(path)
data.head(20)
```

```python
# Vypiš všechny řádky se jmény, jejichž nositelé mají průměrný věk vyšší než 60 (hodnota ve sloupci prumerny_vek je větší než 60).
over_60 = data[
    (data["prumerny_vek"] > 60)
]
over_60
```

```python
#Vypiš pouze jména z těch řádků, kde četnost je mezi 80 000 a 100 000.
freq_80_to_100_k = data[
    (data["cetnost"] >= 80000) & (data["cetnost"] <= 100000)
]
freq_80_to_100_k
```

```python
#Vypiš jména a četnost pro jména se slovanským nebo hebrejským původem. Kolik takových jmen je?
slav_hebrew_names = data[
    (data["puvod"].str.lower().isin(["slovanský", "hebrejský"]))
]
slav_hebrew_names
```
