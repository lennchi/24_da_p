Problémem dnešních potravin je často nedostatek vlákniny. Konzumace potravin bohatých na vlákninu může být pro řadu lidí zdravotně prospěná. Vyhledej v tabulce food_nutrient potraviny, které obsahují alespoň 10 gramů vlákniny. Vláknina je uložená pod názvem Fiber, total dietary. Napiš dotaz jako jeden příkaz a využij operátor &.

```python
import pandas as pd
path = "https://kodim.cz/cms/assets/czechitas/python-data-1/python-pro-data-1/podmineny-vyber/podmineny-vyber/food_nutrient.csv"
data = pd.read_csv(path)
data.head(20)
```

```python
# Which items have at least 10g of fiber
enough_fiber = data[
    (data["name"] == "Fiber, total dietary")
    & (data["amount"] >= 10)
]
enough_fiber
```

```python
# Which rows contain > 100 000 mg
# Which rows contain > 100 g
erroneous_amount = data[
    (data["amount"] > 100000)
    & (data["unit_name"].str.lower() == "mg")
    |
    (data["amount"] > 100)
    & (data["unit_name"].str.lower() == "g")
]
erroneous_amount
```
